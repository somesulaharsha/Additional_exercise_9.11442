```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
iris=sns.load_dataset("iris")
```

```python
iris.head(5)
```

# Scatterplot

```python
plt.scatter(iris["petal_length"],iris["petal_width"], color ="red")
plt.title("scatter plot for bivariate analysis")
plt.xlabel("petal length")
plt.ylabel("petal width")
plt.show()
```

# Histogram


```python
plt.hist(iris["sepal_width"],bins = 20)
plt.title("Histogram for sepal width")
plt.xlabel("sepal width")
plt.ylabel("Frequency")
plt.show()
```

# Bar Chart:

```python
sns.barplot(iris["species"], iris["sepal_length"])
plt.title("Barplot")
```

# Countplot:

```python
sns.countplot(x='species',data=iris)
plt.title("Countplot for species")
```

# Boxplot:

```python
sns.boxplot(x="sepal_length", data=iris)
plt.title("Boxplot showing distribution for sepal length")
```

```python
sns.boxplot(x= "species", y= "sepal_length", data=iris)
plt.title("Boxplot for coparision across levels of species")
```

```python
sns.boxplot(data=iris, orient='h')
plt.title("Boxplot for entire dataset")
```

# Heatmap:

```python
sns.heatmap(iris.corr(),cmap='coolwarm',annot=True)
```

# Distplot:

```python
sns.distplot(iris['sepal_width'],bins = 25,kde= True)
```

# Jointplot:

```python
sns.jointplot(x='sepal_length',y='petal_length',data=iris,kind='scatter')
```
